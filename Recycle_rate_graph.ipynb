```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Recycling_rate1.csv", sep=",")
# rydde litt
df.columns = [c.strip() for c in df.columns]
for col in ["Recycling rate (%)","EU Target (%)"]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# sorter lavest øverst (pent i horisontal graf)
df = df.sort_values("Recycling rate (%)")

ax = df.plot(
    kind="barh",
    x="EEE Category",
    y=["Recycling rate (%)","EU Target (%)"],
    figsize=(10,6)
)
ax.set_title("Recycling rate vs EU WEEE targets (2023–2025)")
ax.set_xlabel("Percent (%)")
ax.legend(title="")
plt.tight_layout()
plt.show()
```
